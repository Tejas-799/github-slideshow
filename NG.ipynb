```python
#23.write a python program to display student marks from the record.
student_marks = {
    "vinay": 69,
    "tejas": 99,
    "sachin": 50,
    "Nirmal": 99,
    "sekh": 8
}
def display_student_marks(record):
    print("Student Marks:")
    for student, marks in record.items():
        print(f"{student}: {marks}")
display_student_marks(student_marks)
```

```python
#24.write a python to create a labeled bar graph using matplotlib.pyplot
import matplotlib.pyplot as plt
categories = ['vinay', 'nirmal', 'sachin', 'tejas']
values = [5, 30, 5, 30]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Labeled Bar Graph')
plt.show()
```

```python
#25.writeapythonprogramtocreatealabeledpiegraphusing mathplotlib.pyplot
import matplotlib.pyplot as plt
labels = ['vinay', 'nirmal', 'sachin', 'tejas']
sizes = [5, 45, 5, 45]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Labeled Pie Chart')
plt.axis('equal')
plt.show()

```
